```python
import os
base = 'Year Wise'
print()

# Check if the directory exists
if not os.path.exists(base):
    # Create the directory if it doesn't exist
    os.mkdir(base)
    print(f"Directory '{base}' created.")

for year in np.arange(2001,2012):
    print(f'YEAR {year}:',df[df['Year']==year].shape[0],'cases')
    temp=df[df['Year']==year]
    path = os.path.join(base,f"Indian Suicide Statistics {year}.csv")
#     path=f'.\\CountryWise\\{i}.csv'
    temp.to_csv(path,index=False)
```
